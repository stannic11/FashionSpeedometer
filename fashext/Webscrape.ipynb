```python
import requests
from bs4 import BeautifulSoup

urlFromExtension = "https://www.uniqlo.com/us/en/women-plusj-supima-cotton-oversized-half-sleeve-t-shirt-437838.html?dwvar_437838_color=COL08&cgid=women-plusj#start=1&cgid=women-plusj"
res = requests.get(urlFromExtension)
displayfileA = open("fileA.txt","w")
displayfileB = open("fileB.txt","w")
displayfileC = open("fileC.txt","w")
displayfileD = open("fileD.txt","w")
BioKeywords = ["polyester","nylon","cotton", "rayon", "cellulose", "hemp", "wool"]
FPKeywords=["organic", "wool","fur", "leather","polyester","nylon","cotton" "recycled"]
BioKeywordsFound="false"
FPKeywordsFound="false"
soup = BeautifulSoup(res.text, 'html.parser')

for x in BioKeywords:
    if x in res.text.lower():
        if (BioKeywords.index(x)<2) and (BioKeywords.index(x)>-1):
            print (BioKeywords.index(x))
            displayfileA.write("This product contains synthetic fibers. Synthetic fibers do not naturally degrade, meaning that this piece of clothing will remain in a landfill long after its use. Additionally, synthetic fibers release plastic microfibers when washed, which pollute our water systems and wreak havoc on marine life.\n")
            displayfileA.write("\n")
            displayfileA.write("Score: 1")
            BioKeywordsFound="true"
        elif (BioKeywords.index(x)<7) and (BioKeywords.index(x)>1):
            print (BioKeywords.index(x))
            displayfileA.write("This product contains organic fabrics. These materials are very eco-friendly and will naturally degrade after disposal.\n")
            displayfileA.write("\n")
            BioKeywordsFound="true"
            displayfileA.write("Score: 5")
if (BioKeywordsFound=="false"):
    displayfileA.write("Not enough information found.")
    displayfileA.write("Score: 2")
            
for x in FPKeywords:
    if x in res.text.lower():
        if (FPKeywords.index(x)==0):
            print (FPKeywords.index(x))
            displayfileB.write("This product likely contains synthetic dyes, which require a resource-intensive manufacturing process. Producing 1 ton of fabric with synthetic dye requires over 200 tons of water.\n")
            displayfileB.write("\n")
            displayfileB.write("Score: 1")
            FPKeywordsFound="true"
        elif (FPKeywords.index(x)<4) and (FPKeywords.index(x)>0):
            print (FPKeywords.index(x))
            displayfileB.write("This product contains animal products. Large scale livestock operations required to produce animal fabrics are responsible for huge methane outputs. Additionally, tanning and preservation processes require carcinogenic chemicals that risk environmental safety if disposed improperly.\n")
            displayfileB.write("\n")
            displayfileB.write("Score: 3")
            FPKeywordsFound="true"
        elif (FPKeywords.index(x)==4):
            print (FPKeywords.index(x))
            displayfileB.write("This product contains cotton. Although biodegradable, cotton is one of the most pesticide-heavy crops. Not only do these pesticide treatments pollute water systems, depleting the safe water supply of humans and animals alike, they significantly degrade soil quality, preventing future crops from growing.\n")
            displayfileB.write("\n")
            displayfileB.write("Score: 2")
            FPKeywordsFound="true"
        elif (FPKeywords.index(x)<6) and (FPKeywords.index(x)>4):
            print (FPKeywords.index(x))
            displayfileB.write("This product contains synthetic fibers. Synthetic fibers are produced from petroleum, which means they depend on fossil fuel extraction. Fossil fuel extraction brings a whole host of problems from drinking water contamination, to oil spills, to methane emissions.\n")
            displayfileB.write("\n")
            displayfileB.write("Score: 1")
            FPKeywordsFound="true"
        elif (FPKeywords.index(x)==6):
            print (FPKeywords.index(x))
            displayfileB.write("This product contains recycled materials, suggesting that its production involved a smaller carbon footprint.\n")
            displayfileB.write("\n")
            displayfileB.write("Score: 5")
            FPKeywordsFound="true"
if (FPKeywordsFound=="false"):
    displayfileB.write("Not enough information found.\n") 
    displayfileB.write("Score: 2")
            
if ("fair trade" in res.text.lower()):
    displayfileC.write("This product is Fair Trade Certified, meaning that it supports equitable trade relationships that provide fair and safe working conditions./n")
    displayfileC.write("\n")
    displayfileC.write("Score: 5")
else:
    displayfileC.write("This product is not Fair Trade Certified; the equity of trade relationships cannot be determined. /n")
    displayfileC.write("\n")
    displayfileC.write("Score: 0")
UnethicalBrands = requests.get("https://livefairethical.wordpress.com/the-sweatshop-list-2/")
currentbrandStr=soup.title.string
brandIndex=currentbrandStr.rfind("|")
brand=currentbrandStr[brandIndex+1:len(currentbrandStr)-1]
print(brand)
if brand.strip() in UnethicalBrands.text:
    displayfileD.write("This brand has serious allegations of using sweatshops to produce its products. Sweatshops rely on gross exploitation of impoverished workers, especially children, in developing countries./n")
    displayfileD.write("\n")
    displayfileD.write("Score: 0")
else:
    displayfileD.write("Not enough information. \n")
    displayfileD.write("\n")
    displayfileD.write("Score: 2")
```
